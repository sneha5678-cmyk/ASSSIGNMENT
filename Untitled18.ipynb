```python
# Install necessary libraries
!pip install pandas seaborn matplotlib

# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/listings_detailed.csv'  # Your file path
df = pd.read_csv(file_path)

# Data cleaning and preprocessing
# Check for missing values and handle them
df.isnull().sum()

# Fill missing values or drop rows with missing values
df = df.dropna(subset=['price', 'room_type', 'number_of_reviews'])

# Ensure 'price' is numeric
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Convert 'last_scraped' to datetime
df['last_scraped'] = pd.to_datetime(df['last_scraped'], errors='coerce')

# Check data types after cleaning
print(df.dtypes)

# Exploratory Data Analysis (EDA)

# Plot price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()

# Plot number of reviews distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['number_of_reviews'], bins=30, kde=True)
plt.title('Number of Reviews Distribution')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.show()

# Room type distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='room_type', data=df)
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()

# Correlation matrix to understand relationships
corr_matrix = df[['price', 'number_of_reviews', 'availability_30', 'availability_60', 'availability_365']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Average price by room type
avg_price_by_room_type = df.groupby('room_type')['price'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='room_type', y='price', data=avg_price_by_room_type)
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price ($)')
plt.show()

# Availability vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='availability_365', y='price', data=df)
plt.title('Availability vs Price')
plt.xlabel('Availability 365')
plt.ylabel('Price ($)')
plt.show()

# Review scores and price relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='review_scores_rating', y='price', data=df)
plt.title('Review Scores Rating vs Price')
plt.xlabel('Review Scores Rating')
plt.ylabel('Price ($)')
plt.show()

# Host's response time vs Price
plt.figure(figsize=(10, 6))
sns.boxplot(x='host_response_time', y='price', data=df)
plt.title('Host Response Time vs Price')
plt.xlabel('Host Response Time')
plt.ylabel('Price ($)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('/content/listings_detailed.csv')

# Basic Data Preprocessing
df.dropna(subset=['price'], inplace=True)
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Convert categorical columns like 'room_type' into numerical values
le = LabelEncoder()
df['room_type'] = le.fit_transform(df['room_type'])

# Drop irrelevant columns
df.drop(columns=['listing_url', 'scrape_id', 'last_scraped', 'host_name'], inplace=True)

# Analysis 1: Amenities vs. Price
df['amenities_list'] = df['amenities'].apply(lambda x: x.split(','))
amenities_counts = pd.Series([amenity for sublist in df['amenities_list'] for amenity in sublist]).value_counts()

plt.figure(figsize=(12, 6))
amenities_counts.head(20).plot(kind='barh', color='skyblue')
plt.title('Top 20 Most Common Amenities')
plt.xlabel('Count')
plt.ylabel('Amenity')
plt.show()

# Analysis 2: Price by Location (Map)
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=10)
for i, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"Price: ${row['price']}"
    ).add_to(m)
m

# Analysis 3: Review Scores vs. Price
review_scores = ['review_scores_rating', 'review_scores_cleanliness', 'review_scores_location', 'review_scores_communication']
corr = df[review_scores + ['price']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Between Review Scores and Price')
plt.show()

# Analysis 4: Property Type and Room Type Comparison
plt.figure(figsize=(10, 6))
sns.boxplot(x='property_type', y='price', data=df)
plt.title('Price Distribution by Property Type')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price Distribution by Room Type')
plt.show()

# Analysis 5: Host Characteristics vs. Price
plt.figure(figsize=(10, 6))
sns.boxplot(x='host_location', y='price', data=df)
plt.title('Price Distribution by Host Location')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='price', data=df)
plt.title('Price vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
# Sample a smaller subset of the dataset (let's say 1000 random rows)
sample_df = df.sample(n=1000, random_state=42)

# Analysis: Host Characteristics vs. Price (using the smaller sample)
plt.figure(figsize=(10, 6))
sns.boxplot(x='host_location', y='price', data=sample_df)
plt.title('Price Distribution by Host Location (Sampled Data)')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Visualize Price Distribution by Host Location
plt.figure(figsize=(12, 6))
sns.boxplot(x='host_location', y='price', data=sample_df)
plt.title("Price Distribution by Host Location")
plt.xticks(rotation=90)
plt.show()

# 2. Top 5 Host Locations with Highest Average Price
top_5_locations = sample_df.groupby('host_location')['price'].mean().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
top_5_locations.plot(kind='bar', color='skyblue')
plt.title("Top 5 Host Locations with Highest Average Price")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()

# 3. Property Type with the Highest Average Price
avg_price_property_type = sample_df.groupby('property_type')['price'].mean().sort_values(ascending=False).head(1)
plt.figure(figsize=(6, 6))
avg_price_property_type.plot(kind='bar', color='lightcoral')
plt.title("Property Type with the Highest Average Price")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()

# 4. Room Type with the Highest Price Distribution Variation
room_type_variation = sample_df.groupby('room_type')['price'].std().sort_values(ascending=False).head(1)
plt.figure(figsize=(6, 6))
room_type_variation.plot(kind='bar', color='gold')
plt.title("Room Type with the Highest Price Distribution Variation")
plt.ylabel("Price Standard Deviation")
plt.xticks(rotation=45)
plt.show()

# 5. Correlation Between Review Scores and Price
plt.figure(figsize=(8, 6))
sns.scatterplot(x='review_scores_rating', y='price', data=sample_df)
plt.title("Correlation Between Review Scores and Price")
plt.xlabel("Review Scores Rating")
plt.ylabel("Price")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = '/content/listings_detailed.csv'
df = pd.read_csv(file_path)

# Check for missing values and clean data
df = df.dropna(subset=['host_name'])

# Count how many properties each host owns
host_property_count = df['host_name'].value_counts()

# Set a threshold to categorize hosts who own multiple properties (e.g. more than 1 property)
multi_property_owners = host_property_count[host_property_count > 1]

# Set up visualizations
plt.figure(figsize=(12, 6))

# Plot the distribution of number of properties owned by each host
sns.histplot(host_property_count, bins=20, kde=True, color='skyblue')
plt.axvline(x=1, color='red', linestyle='--', label='Single Property Owners')
plt.title('Distribution of Properties Owned by Hosts')
plt.xlabel('Number of Properties Owned')
plt.ylabel('Number of Hosts')
plt.legend()
plt.show()

# Plot the number of multi-property owners
plt.figure(figsize=(8, 6))
sns.countplot(x=multi_property_owners.index, order=multi_property_owners.index, color='lightgreen')
plt.title('Hosts Who Own Multiple Properties')
plt.xlabel('Host Name')
plt.ylabel('Number of Properties Owned')
plt.xticks(rotation=90)
plt.show()

# Summary of multi-property owners
print("Summary of Multi-Property Owners:")
print(multi_property_owners.head())
```
